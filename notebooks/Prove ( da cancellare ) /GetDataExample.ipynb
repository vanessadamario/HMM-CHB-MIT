```python
with open('Dataset1.pickle','rb') as f:
    [Dataset1] = pickle.load(f)
```

```python
print(Dataset1)
```

```python
Dates = Dataset1['Trade_Date'].values
Dataset1 = Dataset1.drop(columns=['Trade_Date'])
Data = Dataset1.values
```

```python
plt.rcParams['figure.figsize'] = [15,10]
for i in range(np.size(Data,axis=1)):
    plt.plot(Dates,Data[:,i],label = Dataset1.columns[i])
plt.legend()
plt.show()
```

```python
with open('Dataset3.pickle','rb') as f:
    [Dataset3] = pickle.load(f)
print(Dataset3)
```

```python
np.corrcoef(Dataset3['M3'].values, Dataset3['M2'].values)
```

```python
Dates3 = Dataset3['Trade_Date'].values
Dataset3 = Dataset3.drop(columns=['Trade_Date'])
Data3 = Dataset3.values
```

```python
p_vec_0 = np.random.uniform(30,100,Data3.shape[1])
Data = np.zeros((np.size(Data3,axis=0)+1,Data3.shape[1]))
Data[0, :] = p_vec_0
for j in range(1, Data3.shape[0]+1):
    Data[j,:] =  Data[j-1,:] + Data3[j-1,:]*Data[j-1,:]/100

for i in range(Data.shape[1]):
    plt.plot(Data[:, i], label = Dataset3.columns[i])
plt.legend(ncol=3) 
plt.show()
```
