```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
with open('Cluster_methods_comparison.pickle', 'rb') as f:
   [trial] =  pickle.load(f)
print(trial)
```

```python
clustering_methods = []
Precision_generator = []
dim_TS = []
N_states = []

for i in trial.keys():
    N_states.append(i[0])
    dim_TS.append(i[1])
    Precision_generator.append(i[2])
    clustering_methods.append(i[3])

N_states = np.sort(list(set(N_states)))
dim_TS = np.sort(list(set(dim_TS)))
Precision_generator = list(set(Precision_generator))
clustering_methods = list(set(clustering_methods))
```

```python
# V-measure regain-random

plt.figure(figsize=(10,10))
for meth in clustering_methods:
    temp = []
    x_tick = []
    for i in N_states:
        for j in dim_TS:
            temp.append(trial[(i,j,Precision_generator[0],meth)]['v_measure [0, 1]'])
            x_tick.append('('+str(i)+','+str(j)+')')
    plt.plot(temp, label = meth)
    
plt.xticks(np.arange(len(x_tick)),x_tick,rotation=90)
plt.title('V-measure with precision drawn from '+ str(Precision_generator[0]) )
plt.xlabel('(N_states,N_dim)')
plt.ylabel('V-measure')
plt.legend()
plt.show()        
```

```python
 # V-measure regain

plt.figure(figsize=(10,10))
for meth in clustering_methods:
    temp = []
    x_tick = []
    for i in N_states:
        for j in dim_TS:
            temp.append(trial[(i,j,Precision_generator[1],meth)]['v_measure [0, 1]'])
            x_tick.append('('+str(i)+','+str(j)+')')
    plt.plot(temp, label = meth)
    
plt.xticks(np.arange(len(x_tick)),x_tick,rotation=90)
plt.title('V-measure with precision drawn from '+ str(Precision_generator[1]) )
plt.xlabel('(N_states,N_dim)')
plt.ylabel('V-measure')
plt.legend()
plt.show() 
```

```python
# MCC regain-random

plt.figure(figsize=(10,10))
for meth in clustering_methods:
    temp = []
    x_tick = []
    for i in N_states:
        for j in dim_TS:
            temp.append(trial[(i,j,Precision_generator[0],meth)]['weighted_mean_mcc [-1, 1]'])
            x_tick.append('('+str(i)+','+str(j)+')')
    plt.plot(temp, label = meth)
    
plt.xticks(np.arange(len(x_tick)),x_tick,rotation=90)
plt.title('MCC with precision drawn from '+ str(Precision_generator[0]) )
plt.xlabel('(N_states,N_dim)')
plt.ylabel('MCC')
plt.legend()
plt.show() 
```
